```python
df_country = pd.read_excel('/content/Country-Code.xlsx')
```

```python
#let us try to combine the two data frames on the basis of country code

combined_df = pd.merge(df, df_country, on = 'Country Code', how ='left')
```

```python
combined_df.head()
```

```python
#check datatypes
combined_df.dtypes
```

```python
combined_df['Country'].value_counts()
#we can see that more than 80-85 percent outlets are in India
```

```python
country_names  = combined_df['Country'].value_counts().index
country_names
```

```python
country_values = combined_df['Country'].value_counts().values
```

```python
#Pie Chart

plt.pie(country_values[0:3], labels = country_names[0:3], autopct='%1.2f%%')
```

```python
combined_df.columns
```

```python
ratings = combined_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
```

```python
ratings
```

```python
#Observations:
# 1. For rating = 4.5-4.9, text = "excellent"
# 2. For rating = 4.0-4.4, text = "Very Good"
# 3. For rating = 3.5-3.9, text = "Good"
# 4. For rating = 2.5-3.4, text = "Average"
# 5. For rating = 1.8-2.4, text = "Poor"

# A lot of people haven't given rating and hence of 0 ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data = ratings, hue='Rating color', palette=['black', 'red', 'orange','yellow','green', 'green'])
```

```python
#observations

# 1. Not rated count is very high
# 2. Max ratings are between 2.7 - 4.2
# 3. If there is a rating that is missing we can impute the most frequent rating
```

```python
sns.countplot(x="Rating color", data = ratings,  palette=['black', 'red', 'orange','yellow','green', 'green'])
```
